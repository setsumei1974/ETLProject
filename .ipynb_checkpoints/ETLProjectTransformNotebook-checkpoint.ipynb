```python
import pandas as pd
from pandas import DataFrame
from sqlalchemy import create_engine
```

```python
CoffeeProducers_df = pd.read_csv("CoffeeProductionConsumption.csv")
CoffeeProducers_df.head()
```

```python
CoffeeProducers = CoffeeProducers_df[["COUNTRY", 'YEAR', "TOTAL_PRODUCTION", "EXPORTABLE_PRODUCTION"]].copy()
CoffeeProducers.head()
```

```python
CoffeeProducers2 = CoffeeProducers.rename(columns={"COUNTRY": "country", "YEAR": "year", "TOTAL_PRODUCTION": "totalproduction", "EXPORTABLE_PRODUCTION": "exportableproduction"})
CoffeeProducers2.head()
#CoffeeProducers2_df.dtypes
```

```python
CoffeeProducers2.to_csv("CoffeeProductionConsumptionRevised.csv")
```

```python
StarbucksLocations_df = pd.read_csv("StarbucksLocations.csv")
StarbucksLocations_df.head()
```

```python
StarbucksLocations = StarbucksLocations_df[["Store Number", 'City', "State/Province", "Country"]].copy()
StarbucksLocations.head()
```

```python
StarbucksLocations2 = StarbucksLocations.rename(columns={"Store Number": "store", "City": "city", "State/Province": "stateprovince", "Country": "country"})
StarbucksLocations2.head()
#StarbucksLocations2.dtypes
```

```python
StarbucksLocations2.to_csv("StarbucksLocationsRevised.csv")
```

```python
ConversionCode_df = pd.read_csv("CountrytoAlpha2.csv")
ConversionCode_df.head()
```

```python
ConversionCode = ConversionCode_df[["Country", "Alpha2"]].copy()
ConversionCode.head()
```

```python
ConversionCode2 = ConversionCode.rename(columns={"Country": "country", "Alpha2": "alpha2"})
ConversionCode2.head()
#ConversionCode2.dtypes
```

```python
ConversionCode2.to_csv("CountrytoAlpha2Revised.csv")
```

```python
rds_connection_string = "postgres:<password>@localhost:5432/coffee_db"
engine = create_engine(f"postgresql://{rds_connection_string}")
```

```python
engine.table_names()
```

```python
CoffeeProducers2.to_sql(name="coffee_production", con=engine, if_exists="append", index=False)
StarbucksLocations2.to_sql(name="starbucks_locations", con=engine, if_exists="append", index=False)
ConversionCode2.to_sql(name="conversion_table", con=engine, if_exists="append", index=False)
```

```python
pd.read_sql_query("select * from coffee_production", con=engine).head()
```

```python
pd.read_sql_query("select * from starbucks_locations", con=engine).head()
```

```python
pd.read_sql_query("select * from conversion_table", con=engine).head()
```

```python
SAProductionLocations = pd.read_sql_query("SELECT coffee_production.country, coffee_production.year, coffee_production.totalproduction, coffee_production.exportableproduction, starbucks_locations.store, starbucks_locations.city FROM coffee_production \
JOIN conversion_table ON coffee_production.country = conversion_table.country \
JOIN starbucks_locations ON conversion_table.alpha2 = starbucks_locations.country;", con=engine)

SAProductionLocations.head()
```

```python
Brazil = SAProductionLocations.loc[SAProductionLocations["country"] == "Brazil"]

Brazil
```

```python
Bolivia = SAProductionLocations.loc[SAProductionLocations["country"] == "Bolivia"]

Bolivia
```
