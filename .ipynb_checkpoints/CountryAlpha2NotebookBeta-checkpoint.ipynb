```python
import pandas as pd
import numpy as np
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as bs
import requests
```

```python
req = Request('https://www.nationsonline.org/oneworld/country_code_list.htm', headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
webpage
```

```python
type(webpage)
```

```python
soup = bs(webpage, 'html.parser')
type(soup)
```

```python
with open("CountryCodes.html", "w", encoding='utf-8') as file:
    file.write(str(soup))
```

```python
CountryAlpha2 = pd.read_html("CountryCodes.html")[0]
CountryAlpha2.head()
```

```python
CountryAlpha2_df = CountryAlpha2.drop(CountryAlpha2.columns[[0, 3, 4]], axis=1)
CountryAlpha2_df.head()
```

```python
CountryAlpha2_df2 = CountryAlpha2_df.drop(CountryAlpha2_df.index[0])
CountryAlpha2_df2.head()
```

```python
CountryAlpha2_df3 = CountryAlpha2_df2.rename(columns = {1:"Country", 2:"Alpha2"})
CountryAlpha2_df3.head()
```

```python
CountryAlpha2_df3.to_csv("CountrytoAlpha2.csv")
```
